```
#Paralel olmayan sıradan python veri yapısı ile 
sayilar = [1,2,3,4,5,6,7,8]
kareler = []
for sayi in sayilar:
  kare = sayi * sayi
  kareler.append(kare)
print(sayilar)
print(kareler)
```

```
#RDD --> Resilient Distributed Dataset
#RDD (Dağıtık Veri Yapısı kullanarak) paralel işlem yapıyoruz
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
sayilarRDD = sc.parallelize([1,2,3,4,5,6,7,8])
def kareAl(sayi):
  return (sayi, sayi*sayi , sayi*sayi*sayi )
karelerRDD = sayilarRDD.map(lambda x : kareAl(x))
print(karelerRDD.collect())
```

```
#Lineer Model
def evFiyatTahmini(oda,salon,alan):
  return 1000*oda + 5000*salon + alan * 5000


evFiyatTahmini(3,2,300)
```

```
#Non-Lineer Model (Tree Based Model)
def evFiyatTahmini2(oda,salon,alan):
  if ( oda > 3 ):
    if ( salon > 5):
      if ( alan > 300):
        return 780000
      else:
        return 600000
    else:
      if (alan < 200):
        return 400000
      else:
        return 400000
  else:
    return 200000

  return 1000*oda + 5000*salon + alan * 5000


evFiyatTahmini2(5,2,600)
```
