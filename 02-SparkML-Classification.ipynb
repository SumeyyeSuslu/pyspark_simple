```
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler,StringIndexer
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

spark = SparkSession.builder.appName("Spark ML").getOrCreate()

irisDF = spark.read.option("inferSchema","true").option("header","true").csv("/content/drive/MyDrive/EuroCC/WinterSchool/iris-dataset.txt")

vec  = VectorAssembler(inputCols=irisDF.columns[0:-1],
                       outputCol="features")

irisDF =vec.transform(irisDF)
indxr = StringIndexer(inputCol="class",outputCol="label")
irisDF = indxr.fit(irisDF).transform(irisDF)
irisDF = irisDF.select("features","label")
rfClassifier = RandomForestClassifier()
trainDF,testDF = irisDF.randomSplit([0.75,0.25],seed=123)
model = rfClassifier.fit(trainDF)
sonuclarDF = model.transform(testDF)
#sonuclarDF.show()
eva = MulticlassClassificationEvaluator(metricName="accuracy")
basari = eva.evaluate(sonuclarDF)
print("Başarı oranımız : ",basari)
```
